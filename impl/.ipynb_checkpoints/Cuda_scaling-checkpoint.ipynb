## one vertex per thread (complete)

Simulate the Congest clique model as one vertex per machine. The message sending/receiving is reading and writing to the buffer space. Number of iteration is the number of copying the memory from host to device before the kernel and device to host after the kernel.

less iteration -> less mem copying. But mem copying doesn't take the dominant amount of the time.
When simulating on GPU, the model for time does not work well. 

## Why kernel time is dominat why kernel 2 time composition goes down

1. memory pattern 
transposing the buffer matrix will make reading faster for both kernel 1 and 2 but for complete graph only
2. sorting, computation power on each thread
bottle neck of the whole algo. A cluster leader is in charge of sorting all the candidate edges in the cluster. Sorting in GPU thread isn't optimized for its memory pattern and there are a lot of thread wasted as they are not cluster leader. Sinmulating more vertice on a single thread has higher chance to average the computation for sorting for each thread. And that is the reason of kernel 2 time composition drops as we start to simulating more vertices on a single thread.

## Sparse graph inefficiency
When checking min in coming edges with cluster leaders. The size of buffer needed is at most $O(n)$ for each cluster leader with loop up table which brings $O(n^2)$ to the entire buffer space. So when checking the min edges, we need to have a nested loop to loop up added edges to determine the min.


```C++
for (int i = v_indices[from_v]; i < v_indices[from_v + 1]; ++i) {
    ClusterEdge e = to_cluster_buf[i];
    if (get_cluster_leader(cluster_ids, e.to_v) == to_v) {
        edge = e;
        break;
    }
}
```

````python
def render_strong_scale_sparse(n, avg_rank):
    df = pd.read_csv(cuda_dir(f'strong_scale_sparse_{n}_{avg_rank}/strong_scale_sparse_{n}_{avg_rank}.csv'))

    x = n / df["num_vertex_local"]
    # 8, 16, 32, ..., 128, 256

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, df['total'], label='total time', marker='o')
    plt.plot(x, df['cpu'], label='cpu time', marker='o')

    # kernel time
    plt.plot(x, df['min_from_cluster_kernel'], label='min_from_cluster_kernel time', marker='o')
    plt.plot(x, df['min_to_cluster_kernel'], label='min_to_cluster_kernel time', marker='o')

    # mem copy
    plt.plot(x, df['DtoH'], label='DtoH time', marker='o')
    plt.plot(x, df['HtoD'], label='HtoD time', marker='o')

    plt.xlabel('number of thread')
    plt.ylabel('time (milli-second)')
    plt.title(f'Strong scaling time sparse {n} {avg_rank}')
    plt.xscale('log')

    plt.ylim(bottom=0)
    plt.legend()
    plt.grid(True, which="both", ls="--")
    plt.show()
````

## Diff between MPI and Cuda time composition

The sequential part for cuda: CPU computation, mem copying (allocting the same amount of memory for same graph regardless of the thread count) (which is simulating the message passing in congest clique)

The sequential part for MPI: edge selection (same as CPU computation for CUDA). But the message passing for MPI will scale with more thread as the messages are averaged over the network.

````python
def render_strong_scale_sparse_speedup(n, num_vertex_local, key):
    df = pd.read_csv(cuda_dir(f'strong_scale_sparse_{n}_{num_vertex_local}/strong_scale_sparse_{n}_{num_vertex_local}.csv'))

    x = n / df["num_vertex_local"]
    # 8, 16, 32, ..., 128, 256

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, get_speed_up(list(df[key])[::-1])[::-1], label=f'{key} speed up {n}_{num_vertex_local}', marker='o')
    
    plt.xlabel('number of thread')
    plt.ylabel('speed up')
    plt.title(f'Strong scaling sparse {n} {num_vertex_local} speed up')
    plt.xscale('log')

    plt.ylim(bottom=0)
    plt.legend()
    plt.grid(True, which="both", ls="--")
    
    plt.show()
    
def render_strong_scale_sparse_speedup_array(ns, num_vertex_locals, key):

    # 8, 16, 32, ..., 128, 256

    # Plotting
    plt.figure(figsize=(10, 6))
    for idx in range(len(ns)):
        n = ns[idx]
        num_vertex_local = num_vertex_locals[idx]
        
        df = pd.read_csv(cuda_dir(f'strong_scale_sparse_{n}_{num_vertex_local}/strong_scale_sparse_{n}_{num_vertex_local}.csv'))
        
        x = n / df["num_vertex_local"]
        plt.plot(x, get_speed_up(list(df[key])[::-1])[::-1], label=f'{key} speed up {n}_{num_vertex_local}', marker='o')
    
    plt.xlabel('number of thread')
    plt.ylabel('speed up')
    plt.title(f'Strong scaling sparse speed up')
    plt.xscale('log')

    plt.ylim(bottom=0)
    plt.legend()
    plt.grid(True, which="both", ls="--")
    
    plt.show()
````

````python
# ns = [8192, 8192, 16384, 16384, 32768, 32768]
# num_vertex_locals = [10, 20, 10, 20, 10, 20]
ns = [8192, 16384, 32768, 65536]
num_vertex_locals = [10, 10, 10, 10]
render_strong_scale_sparse_speedup_array(ns, num_vertex_locals, "total")

render_strong_scale_sparse_speedup_array(ns, num_vertex_locals, "min_from_cluster_kernel")


# render_strong_scale_sparse_speedup(8192, 10, "min_from_cluster_kernel")
# render_strong_scale_sparse_speedup(8192, 20, "min_from_cluster_kernel")
# render_strong_scale_complete_speedup(8192, "total")
````
